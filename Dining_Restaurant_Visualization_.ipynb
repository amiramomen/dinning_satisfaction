#### Feature Columns
- Satisfaction: Dining satisfaction level (Satisfaction,dissatisfaction)
- Gender: Gender of the diners (Female, Male)
- Customer Type: The customer type (Loyal customer, disloyal customer)
- Age: The actual age of the diners
- Type of Dining: Purpose of the visit to the restaurant (Personal Dining, Business Dining)
- Ambiance: Dining ambiance in the restaurant (Eco, Eco Plus and Business )
- Distance from Home: Distance traveled to reach the restaurant
- Table Comfort': comfort level of the tables in the dining place
- Reservation Convenience': Satisfaction level of the reservation process (Reservation Process, Waiting Time,etc.)
- Food and Drink Quality': Satisfaction level of the food and drinks
- Seating Location': Satisfaction level of seating locations
- WiFi Service: Satisfaction level of the restaurant's WiFi service
- Entertainment Options: Satisfaction level of available entertainment options (Live Music, Events, Games, etc.)
- Online Reservation Support: Satisfaction level of online reservation support
- Ease of Online Booking : Satisfaction level of the online booking process
- Table Service: Satisfaction level of table service
- Space Between Tables: Satisfaction level of the space between dining tables
- Baggage Handling: Satisfaction level of baggage handling (Coat Check, Personal Items Handling, etc.)
- Check-in Service: Satisfaction level of the check-in service (Reception, Reservation Confirmation, etc.)
- Cleanliness: Satisfaction level of cleanliness in the restaurant
- Online Menu Access: Satisfaction level of online menu access (User-Friendly, Detailed, etc.)
- Wait Time for a Table (Departure Delay in Minutes): Minutes delayed when the diners had to wait for a table


> #### Data used [Dining Satisfaction Dataset](https://github.com/amiramomen/dinning_satisfaction/blob/main/satisfaction.csv)
Take Into consideration that (0: (No Answer / Not Applicable); 1-5 (Actual Rating))

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly
import plotly.express as px
import ipywidgets as widgets
from ipywidgets import interact
```

## 2. <a name="2">Read the dataset</a>
(<a href="#0">Go to top</a>)

```python
df=pd.read_csv('./df_undivided_prepared.csv')
df.head()
```

```python
df.info()
```

## 3. <a name="3">Ask questions </a>



(<a href="#0">Go to top</a>)

```python

category_dropdown = widgets.Dropdown(options=['satisfaction', 'Gender', 'Customer Type', 'Type of Dining','Ambiance'],
                                     value='satisfaction', description='Choose:')

@interact(category=category_dropdown)
def plot_category_distribution(category):

    if category == 'satisfaction':
        sns.countplot(x='satisfaction', data=df, palette=['#24b524', '#cc1d1d'])
        plt.title('Distribution of Satisfaction')
        plt.xlabel('Satisfaction')
        plt.ylabel('Count')
    elif category in ['Gender', 'Customer Type', 'Type of Dining','Ambiance']:
        sns.countplot(x=category, hue='satisfaction', data=df, palette=['#24b524', '#cc1d1d'])
        plt.title(f'Distribution of Satisfaction by {category}')
        plt.xlabel(category)
        plt.ylabel('Count')

    plt.show()
```

```python
sns.set(style="whitegrid")
sns.histplot(data=df, x='Age', bins=30, kde=True, color='green')

plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='satisfaction', y='Distance from Home', data=df, palette=['#24b524', '#cc1d1d'])
plt.title('Box Plot: Satisfaction vs Distance from Home')
plt.xlabel('Satisfaction')
plt.ylabel('Distance from Home')
plt.show()
```

```python
sns.stripplot(x='satisfaction', y='Distance from Home', data=df, palette=['#24b524', '#cc1d1d'], jitter=True)
plt.title('Strip Plot: Satisfaction vs Distance from Home')
plt.xlabel('Satisfaction')
plt.ylabel('Distance from Home')
plt.show()
```

```python
columns_of_interest = ['Table Comfort', 'Reservation Convenience', 'satisfaction']
subset_df = df[columns_of_interest]

correlation_matrix = subset_df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix: Table Comfort, Reservation Convenience, and Overall Satisfaction')
plt.show()
```

```python
columns_of_interest = ['Table Comfort', 'Food and Drink Quality', 'satisfaction']
subset_df = df[columns_of_interest]
correlation_matrix = subset_df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Matrix: Table Comfort, Food and Drink Quality, and Overall Satisfaction')
plt.show()
```

```python
columns_of_interest = ['Reservation Convenience', 'Food and Drink Quality', 'satisfaction']
subset_df = df[columns_of_interest]

# Function to update the plot based on the selected feature
def update_plot(selected_feature):
    plt.figure(figsize=(10, 6))
    subset_df = df[[selected_feature,'Table Comfort']]
    correlation_matrix = subset_df.corr()

    # Create a heatmap
    plt.figure(figsize=(10, 8))
    sns.heatmap(correlation_matrix, annot=True, cmap='RdYlGn', fmt=".2f", linewidths=.5)
    plt.title(f'Correlation Matrix: Table Comfort, {selected_feature}, and Overall Satisfaction')

    plt.show()

# Create a dropdown widget for selecting the feature
feature_selector = widgets.Dropdown(
    options=['Reservation Convenience', 'Food and Drink Quality'],
    value='Reservation Convenience',
    description='Select Feature:',
)

# Create an interactive plot using the update_plot function
interact(update_plot, selected_feature=feature_selector)
```

```python
ratings = df['Food and Drink Quality']

# Create a bar chart
plt.figure(figsize=(12, 6))
sns.countplot(x=ratings, palette=['#47ff95', '#1efc7c', '#2ECC71', '#129B56', '#1D8348', '#186A3B'],order=range(6))
plt.title('Bar Chart: Food and Drink Quality Ratings')
plt.xlabel('Ratings')
plt.ylabel('Count')
plt.show()
```

```python
online_booking_ratings = df['Ease of Online booking']
online_support_ratings = df['Online support']

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(14, 6))

sns.countplot(x=online_booking_ratings, palette=['#47ff95', '#1efc7c', '#2ECC71', '#129B56', '#1D8348', '#186A3B'], ax=axes[0])
axes[0].set_title('Ease of Online Booking Ratings')
axes[0].set_xlabel('Ratings')
axes[0].set_ylabel('Count')

sns.countplot(x=online_support_ratings, palette=['#17ff95', '#1efc7c', '#2ECC71', '#129B56', '#1D8348', '#186A3B'], ax=axes[1])
axes[1].set_title('Online Support Ratings')
axes[1].set_xlabel('Ratings')
axes[1].set_ylabel('Count')

plt.tight_layout()

plt.show()




```

```python

plt.figure(figsize=(12, 6))
sns.countplot(data=df, x=df['Ease of Online booking'], hue=df['Online support'], palette='viridis')
plt.title('Ratings for Ease of Online booking and Online support Tables')
plt.xlabel('Ease of Online booking Ratings')
plt.ylabel('Count')
plt.legend(title='SOnline support Tables')
plt.show()
```

```python

plt.figure(figsize=(12, 6))
sns.countplot(data=df, x=df['Table Service'], hue=df['Space Between Tables'], palette='viridis')
plt.title('Ratings for Table Service and Space Between Tables')
plt.xlabel('Table Service Ratings')
plt.ylabel('Count')
plt.legend(title='Space Between Tables')
plt.show()
```

```python

wait_time = df['Wait Time for a Table']
satisfaction = df['satisfaction']

plt.figure(figsize=(12, 6))

sns.histplot(wait_time, bins=30, kde=False, color='skyblue', alpha=0.7)
sns.histplot(wait_time[satisfaction == 'satisfied'], bins=30, kde=False, color='green', alpha=0.7)
sns.histplot(wait_time[satisfaction == 'dissatisfied'], bins=30, kde=False, color='red', alpha=0.7)



plt.title('Distribution of Wait Time for a Table by Satisfaction Levels')
plt.xlabel('Wait Time for a Table')
plt.ylabel('Count')
plt.legend(['All Customers','Satisfied Customers', 'Dissatisfied Customers'])
plt.show()
```

```python

cleanliness_ratings = df['Cleanliness']
food_drink_ratings = df['Food and Drink Quality']

pie_df = pd.DataFrame({
    'Cleanliness': cleanliness_ratings.value_counts(),
    'Food & Drink Quality': food_drink_ratings.value_counts()
})

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))
axes[0].pie(pie_df['Cleanliness'], labels=pie_df['Cleanliness'].index, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'lightcyan'])
axes[0].set_title('Cleanliness Ratings')
axes[1].pie(pie_df['Food & Drink Quality'], labels=pie_df['Food & Drink Quality'].index, autopct='%1.1f%%', startangle=90, colors=['lightgreen', 'palegreen'])
axes[1].set_title('Food & Drink Quality Ratings')

plt.show()
```

```python

cleanliness_ratings = df['Cleanliness']
food_drink_ratings = df['Food and Drink Quality']

bar_df = pd.crosstab(cleanliness_ratings, food_drink_ratings)

plt.figure(figsize=(10, 8))
bar_df.plot(kind='bar', stacked=True, cmap='viridis')
plt.title('Clustered Bar Chart: Cleanliness vs Food & Drink Quality Ratings')
plt.xlabel('Cleanliness Ratings')
plt.ylabel('Count')
plt.legend(title='Food & Drink Quality Ratings', bbox_to_anchor=(1, 1))
plt.show()
```
